```python
# Create pandas DataFrames
df_khader = pd.DataFrame(data_khader)
df_proposed = pd.DataFrame(data_proposed)

# Columns to normalize
columns_to_normalize = ['blockGen time', 'proofGen time', 'proofVer time']

# Normalize using Z-score
for col in columns_to_normalize:
    mean_khader = df_khader[col].mean()
    std_khader = df_khader[col].std()
    df_khader[col] = (df_khader[col] - mean_khader) / std_khader
    
    mean_proposed = df_proposed[col].mean()
    std_proposed = df_proposed[col].std()
    df_proposed[col] = (df_proposed[col] - mean_proposed) / std_proposed

# Plot the normalized data
plt.figure(figsize=(10, 6))
for col in columns_to_normalize:
    block = (df_khader['file-size'])
    plt.plot(block, df_khader[col], marker='o', label=f'Khader - {col}',linestyle='dashed')
    plt.plot(block, df_proposed[col], marker='o', label=f'Proposed - {col}')

plt.xlabel('Encrypted Blocks of 416 bytes')
plt.ylabel('Time (Z-score) (sec)')
# plt.title('Performance of our proposed scheme and Khedrs on different set of blocks')
plt.legend()
plt.grid(True)
plt.show()
```
